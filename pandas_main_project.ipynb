```python
#44.Create Age_Group for customers (Child/Teen/Adult/Senior).
def age_group(age):
  if age < 13:
    return 'Child'
  elif age < 20:
    return 'Teen'
  elif age < 60:
    return 'Adult'
  else:
    return 'Senior'
df['age_group'] = df ['Age'].apply(age_group)
```

```python
#45.Create Loyalty_Level based on total purchase.
def loyalty_level(total):
  if total < 1000:
    return 'Bronze'
  elif total < 5000:
    return 'Silver'
  elif total < 10000:
    return 'Gold'
  else:
    return 'Platinum'
df ['Loyalty_Level'] = df ['Total_Purchase'].apply(loyalty_level)
```
